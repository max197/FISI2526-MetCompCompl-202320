```python
N1 = 5
a1 = 1 #bueno
N2 = 25
a2 = 5 #bueno
```

```python
probs = np.linspace(0,1,1000)
fig,ax = plt.subplots(figsize=(10,5))

y1 = []
for p in probs:
    obs = np.random.binomial(N1,p,1000)
    exitos = np.where(obs==a1)
    probexito =(exitos[0].shape[0])/1000
    y1.append(probexito)

ax.scatter(probs,y1)

y2 = []
for p in probs:
    obs = np.random.binomial(N2,p,1000)
    exitos = np.where(obs==a2)
    probexito =(exitos[0].shape[0])/1000
    y2.append(probexito)

ax.scatter(probs,y2)
```
