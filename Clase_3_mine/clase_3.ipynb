```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(start=0,stop = 7.5,num =10)
g = -9.8
vinital = 10
H = 201
```

```python
def posicion(x0,v0,g,t):
  
  return x0+v0*t+(0.5*g*(t**2))
```

```python
posicion(H,vinital,g,t)
```

```python
def velocidad(v0,g,t):
    resultado = v0-g*t
    return resultado
```

```python
y = posicion(H,vinital, g, t)
```
