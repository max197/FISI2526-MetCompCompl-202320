# **Punto 1**

### **Punto 1.1**

```python
#Constantes
r = 25/2 #radio bucle
Bo= 0.05 #
f = 7 #frecuencia
omega = 3.5 #velocidad angular
R = 1.75 #resistencia
```

```python
def flujo_magnetico(t,r,Bo,f,omega):
    return np.pi*(r**2)*Bo*sym.cos(omega*t)*sym.cos(2*np.pi*f*t)
```

```python
t = sym.Symbol('t', real =True)
y = flujo_magnetico(t,r,Bo,f,omega)
```

```python
df = sym.diff(y,t,1)
df = sym.lambdify(t,df,'numpy')
```

```python
def corriente_bucle(t,R=1.75):
    return (-1/R)*df(t)
```

```python
t = np.linspace(0,2,1000)
y = corriente_bucle(t,R)
fig, ax = plt.subplots(figsize =(10,5))
ax.scatter(t,y)
ax.set_title(
    "Corriente inducida sobre el bucle\n"+r"$B_o=0.05T,f =7Hz,\Omega = 3.5rad/s,R =1.75k\Omega$"
)
plt.xlabel("Time (t)")
```

### **Punto 1.2**

```python
def Derivative(f,x,h=1e-6):
    return (f(x+h)-f(x-h))/(2*h)
```

```python
def GetNewtonMethod(f,df,xn,itmax=10000,precision=1e-14):
    '''
    Implementacion de Newton Rhapson
    '''
    error = 1.
    it = 0
    
    while error >= precision and it < itmax:
        
        try:
            
            xn1 = xn - f(xn)/df(f,xn)
            
            error = np.abs(f(xn)/df(f,xn))
            
        except ZeroDivisionError:
            print('Zero Division')
            
        xn = xn1
        it += 1
        
    if it == itmax:
        return False
    else:
        return xn
```

```python
def GetAllRoots(Function,x,tolerancia=10):
    
    Roots = np.array([])
    
    for i in x:
        
        root = GetNewtonMethod(Function,Derivative,i)
        
        if root != False:
            
            croot = np.round(root, tolerancia)
            
            if croot not in Roots:
                Roots = np.append(Roots,croot)
        
    Roots.sort()
    
    return Roots
```

```python
x = np.linspace(0,0.2,50)
Roots = GetAllRoots(corriente_bucle,x)
Roots = [i for i in Roots if i>=0]
#primeros tres tiempos donde la corriente es cero
Roots = Roots[0:3]
Roots
```

```python
t = np.linspace(0,0.25,1000)
y = corriente_bucle(t,R)
fig, ax = plt.subplots(figsize =(10,5))
ax.scatter(t,y)
ax.set_title("Primeros tres tiempos donde la corriente es cero")
for value in Roots:
    plt.axvline(x=value, color='cyan', linestyle='--')
    plt.text(value, min(y), f'x={value}', color='red', verticalalignment='center', horizontalalignment='center')
plt.xlabel("Time (t)")
```
